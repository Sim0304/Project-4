```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#path to disaster file
disaster_data = Path("Resources\disaster_info.csv")

#read school file
disaster_df = pd.read_csv(disaster_data)
disaster_df.head(1)
```

```python
# drop unnecessary columns from table
df = disaster_df[['title', 'description', 'startDate', 'endDate', 'lat', 'lon', 'Injuries', 'Deaths',  'regions']]
df.head(1)
```

```python
# create seperate columns to split title data into category and subcategory sections
df['Category'], df['Sub-category'] = df['title'].str.split(' - ', n=1).str
df.head(10)
```

```python
# eliminate the title column to clean up the table

dropped_df = df.drop(columns=['title'])
dropped_df.head(1)
```

```python
#Reorganise table
reorganised_df = dropped_df[['Category', 'Sub-category', 'description', 'startDate', 'endDate', 'lat', 'lon', 'Injuries', 'Deaths',  'regions']]
reorganised_df.head()
```

```python
# Separate data for each state
reorganised_df.loc[:, 'new_column'] = reorganised_df['regions'].str.split(';')
reorganised_df
```

```python
# Explode the new_column column
exploded_results = reorganised_df.explode('new_column')
exploded_results
```

```python
# Drop the regions column and rename new_column to Regions
exploded_results = exploded_results.drop('regions', axis = 1).rename(columns={"new_column" : "Regions",
                                                                              "startDate" : "Start Date",
                                                                              "endDate" : "End Date",
                                                                              "description" : "Description",
                                                                              "lat" : "Latitude",
                                                                              "lon" : "Logitude",
                                                                              "Injuries" : "Number of injuries",
                                                                              "Deaths" : "Number of deaths"})
exploded_results.head(20)
```

```python
# Convert the column to datetime format

exploded_results['Start Date'] = pd.to_datetime(exploded_results['Start Date']).dt.date
exploded_results['End Date'] = pd.to_datetime(exploded_results['End Date']).dt.date
exploded_results.head()
```

```python
# Save to csv file
exploded_results.to_csv('Cleaned_file.csv', index=False)
```
